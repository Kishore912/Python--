```python
import pandas as pd
import numpy as np
import seaborn as se
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
Data = load_breast_cancer()
```

```python
dir(Data)
```

```python
df = pd.DataFrame(Data.data,columns=Data.feature_names)
```

```python
df
```

```python
sd = StandardScaler()
scaled_data = sd.fit_transform(df)
```

```python
scaled_data
```

```python
pp = PCA(n_components=2)
x_pca = pp.fit_transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=Data['target'],cmap='plasma')
plt.xlabel('First Principle Component',size=20)
plt.ylabel('Second Principle Component',size=20)
plt.show()
```
