```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python

plt.figure(figsize=(20,6))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(Data.images[i],cmap=plt.cm.gray)
    title = Data.target[i]
    plt.title(title,fontsize=20)
plt.show()
```

```python
df = pd.DataFrame(Data.data,columns=Data.feature_names)
df
```

```python
X = df
y=Data.target
```

```python
scale = StandardScaler()
scaled_df = scale.fit_transform(X)
scaled_df
```

```python
X_train , X_test , y_train , y_test = train_test_split(scaled_df,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
Lg1 = LogisticRegression()
Lg1.fit(X_train,y_train)
```

```python
Lg1.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_train_pca1 , X_test_pca1 , y_train_pca1 , y_test_pca1 = train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
Lg2 = LogisticRegression(max_iter=1000)
Lg2.fit(X_train_pca1,y_train_pca1)
```

```python
Lg2.score(X_test_pca1,y_test_pca1)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_train_pca2 , X_test_pca2 , y_train_pca2 , y_test_pca2 = train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
Lg3 = LogisticRegression(max_iter=1000)
Lg3.fit(X_train_pca2,y_train_pca2)
```

```python
Lg3.score(X_test_pca2,y_test_pca2)
```
