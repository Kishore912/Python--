```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("D:\kishore\python\income.csv")
```

```python
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.show()
```

```python
km = KMeans(n_clusters=3)
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['clusters'] = y_predicted
df
```

```python
df0 = df[df.clusters==0]
df1 = df[df.clusters==1]
df2 = df[df.clusters==2]
```

```python
plt.scatter(df0.Age,df0['Income($)'],c='red')
plt.scatter(df1.Age,df1['Income($)'],c='green')
plt.scatter(df2.Age,df2['Income($)'],c='blue')
plt.xlabel("AGE")
plt.ylabel("INCOME")
plt.legend()
plt.show()
```

HERE YOU CAN SEE THAT THE DATA POINTS ARE NOT GROUPED PROPERLY THIS IS DUE TO NOT SCALED PROPERLY SO WE USE MINMAX SCALER TO SCALE IT PROPERLY

```python
scaler = MinMaxScaler()
scaler.fit(df[['Age']])
df.Age = scaler.transform(df[['Age']])
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted2 = km.fit_predict(df[['Age','Income($)']])
y_predicted2
```

```python
df.clusters = y_predicted2
df
```

```python
km.cluster_centers_
```

```python
df0 = df[df.clusters==0]
df1 = df[df.clusters==1]
df2 = df[df.clusters==2]
```

```python
plt.scatter(df0.Age,df0['Income($)'],c='red')
plt.scatter(df1.Age,df1['Income($)'],c='green')
plt.scatter(df2.Age,df2['Income($)'],c='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='yellow',marker='*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
```

```python
k_rng  = range(1,10)
sse = []
for i in k_rng:
    km = KMeans(n_clusters=i)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)    # inertia is the keyword use to find the sum of squared error            
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.xlabel('k')
plt.ylabel('sum of squared error')
plt.show()
```
