```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import load_digits
```

```python
df = load_digits()
```

```python
df
```

```python
df.data[0]
```

```python
print(df.data.shape)
print(df.target.shape)
```

```python
plt.figure(figsize=(20,6))
for index,(image,label) in enumerate(zip(df.data[0:5],df.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title(label)
    
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.data , df.target , test_size=0.2 ,random_state=42 )
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

using Linear kernel

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```

Using rbf kernel

```python
rbf_model = SVC(kernel='rbf')
rbf_model.fit(X_train,y_train)
```

```python
rbf_model.score(X_test,y_test)
```

```python
y_predict = rbf_model.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_predict,y_test)
```
