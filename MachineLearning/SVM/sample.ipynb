```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets._samples_generator import make_blobs
```

```python
X,y = make_blobs(n_samples=40,centers=2,random_state=20) #n_samples means craetig 40 rows with two centers means 0,1  , make_blobs is a sklearn library for generating data
classifier = SVC(kernel='linear' , C=1.0)
classifier.fit(X,y)
```

```python
plt.scatter(X[:,0],X[:,1],cmap=plt.cm.Paired , c=y)  # here c=y means specifing color for each datapoints based on there class
plt.show()
```

```python
new_data =[[12,13],[2,3]]
print(classifier.predict(new_data))  # here we are predicting new data so the result shows first one is crocodile and second one is a alligator
```

```python
# plot the decision function
classifier = SVC(kernel='linear' , C=1.0)
classifier.fit(X,y)
plt.scatter(X[:,0],X[:,1],cmap=plt.cm.Paired , c=y)

ax= plt.gca()   #getting the axes limits and storing in ax

xlim = ax.get_xlim() # from that current axes extracting x-axes limits and stores in xlim

ylim = ax.get_ylim() # from that current axes extarcting  y-axes limits and stores in ylim
# create grid to evaluate model

xx = np.linspace(xlim[0], xlim[1], 30) # np.linespace() means generating a evenly spaced numbers for a range xlim[0],xlim[1] are the limits for x-axes

yy = np.linspace(ylim[0], ylim[1], 30) #ylim[0],ylim[1] are the limits of y-axes and 30 evenly spaced points along x-axes and y-axes

YY, XX = np.meshgrid(yy, xx)  # here it creates a grid of coordinates taking 1-D array from yy,xx and creating it to a 2-D array of YY,XX

xy= np.vstack([XX.ravel(), YY.ravel()]).T   # ravel() is used to flatten the 2-D array into a 1-D array and vstack() is to vertically stack these the array and T for transpose the array

Z = classifier.decision_function(xy).reshape(XX.shape) # for the grid points xy it computes a decisionfunction value and stores it in Z and reshapes this 1-D array in the form of XX
# plot decision boundary and margins

ax.contour(XX, YY, Z, colors= 'k', levels=[-1, 0, 1],  # contour the values of Z in the XX and YY coordinates , color = 'k' means black and levels are where the contour lines are drawn 
    alpha=0.5, # alpha is the transperancy of the contour lines
    linestyles=['--', '-', '--'])
# plot support vectors

ax.scatter(classifier. support_vectors_[:, 0],  #now scatter the datapoints 
    classifier.support_vectors_[:, 1], s=100,
    linewidth=1, facecolors= 'none')

plt. show()
```
