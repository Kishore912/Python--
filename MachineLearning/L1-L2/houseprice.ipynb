```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("E:\T-Machine\python\Melbourne_housing_FULL.csv")
dataset.head()
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','CouncilArea','Distance','Bathroom','Bedroom2','Landsize','Car','BuildingArea','Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bathroom','Bedroom2','Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset.head()
```
