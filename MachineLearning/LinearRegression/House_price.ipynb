```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv("E:\T-Machine\python\homeprices.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['area'],df['price'],c='red',marker='*')
plt.xlabel('Area(sq feet)')
plt.ylabel('price($)')
plt.show()
```

```python
model = LinearRegression()
model.fit(df[['area']],df[['price']])
```

```python
model.predict(np.array(3300).reshape(1, -1))   #(1,-1) means single row multiple column   (-1,1) means multiple rows single column
```

```python
model.coef_
```

```python
model.intercept_
```

```python
135.78767123*3300+180616.43835616432   #y=mx+c
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,c='red',marker='*')
plt.plot(df.area,model.predict(df[['area']]),c='blue')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
