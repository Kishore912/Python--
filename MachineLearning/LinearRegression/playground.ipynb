```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create some simple data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# Create a linear regression model and fit it using our data
model = LinearRegression()
model.fit(x, y)

# Get the slope (coefficient) and the intercept of the line
slope = model.coef_
intercept = model.intercept_

# Print the slope and intercept
print('Slope (Coefficient): ', slope)
print('Intercept: ', intercept)

# Predict y values
y_pred = model.predict(x)

# Plot the original data (as points) and the line of best fit
plt.scatter(x, y, color ='b', label='Original data')
plt.plot(x, y_pred, color='r', label='Fitted line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
