```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,'o')
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,marker='D')
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,marker='o',mec='black',ms=20)
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,marker='o',mec='black',ms=20,mfc='green')
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,'o:r')
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,'o--r')
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,'o-.r')
plt.show()
```
