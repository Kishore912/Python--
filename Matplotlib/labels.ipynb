```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints)
plt.title("Overview")
plt.xlabel("time_Per_Second")
plt.xlabel("Distance_Covered")
plt.show()
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints)
font1={'family':'serif','color':'red'}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Overview",loc='left' , fontdict=font1)
plt.xlabel("time_Per_Second",fontdict=font2)
plt.ylabel("Distance_Covered",fontdict=font2)
plt.show()
```
