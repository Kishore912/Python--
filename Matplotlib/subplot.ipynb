```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([8,3,7,4,10])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('sales',loc='center')

x=np.array([1,2,3,4,5])
y=np.array([8,3,7,4,10])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('profit',loc='left')

x=np.array([1,2,3,4,5])
y=np.array([8,3,7,4,10])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('expenses',loc='left')

x=np.array([1,2,3,4,5])
y=np.array([8,3,7,4,10])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('loss',loc='left')


plt.suptitle("Bakery")
plt.show()
```
