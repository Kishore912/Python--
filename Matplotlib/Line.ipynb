```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xPoints=np.array([1,2,3,4])
yPoints = np.array([10,6,8,10])
plt.plot(xPoints,yPoints,color='grey',ls='dashed',linewidth=20)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1,y1,x2,y2)
plt.show()
```
