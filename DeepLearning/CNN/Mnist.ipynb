```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.matshow(X_train[0])
plt.xlabel(f"label : {y_train[0]}")
plt.show()
```

```python
plt.matshow(X_train[7])
plt.xlabel(f"label : {y_train[7]}")
plt.show()
```

```python
X_train[0]
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
X_train = np.reshape(X_train, (-1, 28, 28, 1))
```

```python
from tensorflow.keras import models
cnn = models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Conv2D(filters=32,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
cnn.fit(X_train,y_train,epochs=10)
```

```python
cnn.evaluate(X_test,y_test)
```

```python
y_test[0:5]
```

```python
y_pred = cnn.predict(X_test)
y_pred_class = [np.argmax(i) for i in y_pred]
y_pred_class[0:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_class)
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
def actual_value(a):
    plt.figure(figsize=(20,6))
    plt.matshow(X_test[a])
    plt.xlabel(f"actual value:{y_test[a]} predicted value {y_pred_class[a]}")
    plt.show()
    
actual_value(int(input("Enter a number : ")))    
```

```python
import matplotlib.pyplot as plt

def actual_value(a):
    fig, axes = plt.subplots(1, 2, figsize=(20, 6))  # Create a figure with 1 row and 2 columns

    # Plot the actual image with the actual value label
    axes[0].imshow(X_test[a], cmap='gray')  # Assuming images are grayscale
    axes[0].set_xlabel(f"Actual value: {y_test[a]}")

    # Plot the predicted image with the predicted value label
    axes[1].imshow(X_test[a], cmap='gray')  # Assuming images are grayscale
    axes[1].set_title(f"Predicted value: {y_pred_class[a]}")  # Use set_title for subplot title

    # Hide axes for the second plot
    axes[1].axis('off')

    plt.show()

# Call the function with the index of the image you want to display
index_to_display = int(input("Enter the index of the image to display: "))
actual_value(index_to_display)
```
